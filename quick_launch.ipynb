# ComfyUI Quick Launch
Downloads pre-built environment from HuggingFace and starts ComfyUI via Cloudflare tunnel.

```python
#@title Configuration
HF_TOKEN = "" #@param {type:"string"}
ENV_REPO = "datsss/comfyui-env" #@param {type:"string"}
MODEL_REPO = "datsss/my-dataset" #@param {type:"string"}
EXTRA_ARGS = "--dont-print-server --force-fp16" #@param {type:"string"}
```

```python
#@title 1. Download & Extract Pre-built Environment
import os, subprocess, tarfile, time
from pathlib import Path
from concurrent.futures import ThreadPoolExecutor

WORKSPACE = Path("/content")
COMFYUI_DIR = WORKSPACE / "ComfyUI"

if HF_TOKEN:
    os.environ["HF_TOKEN"] = HF_TOKEN
auth = HF_TOKEN if HF_TOKEN.strip() else None

# Enable hf_transfer for faster downloads
subprocess.run("pip install -q hf_transfer", shell=True)
os.environ["HF_HUB_ENABLE_HF_TRANSFER"] = "1"

from huggingface_hub import hf_hub_download

def download_and_extract(filename, extract_to):
    print(f"  Downloading {filename}...")
    path = hf_hub_download(
        repo_id=ENV_REPO, filename=filename, repo_type="dataset",
        token=auth, local_dir=str(WORKSPACE),
    )
    print(f"  Extracting {filename}...")
    with tarfile.open(path, "r:*") as tar:
        tar.extractall(path=extract_to, filter="data")
    print(f"  {filename} done")

print("Downloading pre-built environment...")
start = time.time()

# Ensure ComfyUI dir exists before parallel extraction
COMFYUI_DIR.mkdir(parents=True, exist_ok=True)
with ThreadPoolExecutor(max_workers=2) as pool:
    futs = [
        pool.submit(download_and_extract, "custom_nodes.tar.gz", str(WORKSPACE)),
        pool.submit(download_and_extract, "comfyui-env.tar.gz", str(WORKSPACE)),
    ]
    for f in futs:
        f.result()

elapsed = time.time() - start
print(f"\nEnvironment ready in {elapsed:.0f}s")
```

```python
#@title 2. Restore Python Environment
import subprocess, site
from pathlib import Path

WORKSPACE = Path("/content")
site_packages = Path(site.getsitepackages()[0])

# Restore site-packages
sp_dir = WORKSPACE / "site-packages"
if sp_dir.exists():
    print(f"Restoring site-packages to {site_packages}...")
    # Copy contents into site-packages
    subprocess.run(
        f"cp -rn {sp_dir}/* {site_packages}/ 2>/dev/null; "
        f"cp -rn {sp_dir}/.[!.]* {site_packages}/ 2>/dev/null",
        shell=True
    )
    print("site-packages restored")
else:
    print("No site-packages directory found — skipping")

# Install ComfyUI core deps
comfy_dir = WORKSPACE / "ComfyUI"
req = comfy_dir / "requirements.txt"
if req.exists():
    print("Installing ComfyUI core deps...")
    subprocess.run(f"pip install -q -r {req}", shell=True)

print("Ready!")
```

```python
#@title 3. Launch ComfyUI + Cloudflare Tunnel
import subprocess, threading, re, time, socket

PORT = 8188

def check_server_ready(port, timeout=120):
    start = time.time()
    while time.time() - start < timeout:
        try:
            s = socket.create_connection(("127.0.0.1", port), timeout=1)
            s.close()
            return True
        except:
            time.sleep(0.5)
    return False

def start_tunnel(port):
    # Install cloudflared
    subprocess.run(
        "curl -sL -o /tmp/cf.deb https://github.com/cloudflare/cloudflared/releases/latest/download/cloudflared-linux-amd64.deb",
        shell=True
    )
    subprocess.run("dpkg -i /tmp/cf.deb > /dev/null 2>&1", shell=True)

    p = subprocess.Popen(
        ["cloudflared", "tunnel", "--protocol", "http2", "--url", f"http://127.0.0.1:{port}"],
        stdout=subprocess.PIPE, stderr=subprocess.STDOUT, text=True
    )
    for line in iter(p.stdout.readline, ''):
        if "trycloudflare.com" in line:
            m = re.search(r"https://[a-zA-Z0-9.-]+\.trycloudflare\.com", line)
            if m:
                print(f"\n{'='*50}")
                print(f"Public URL: {m.group(0)}")
                print(f"{'='*50}\n")

# Wait for ComfyUI to start, then launch tunnel
def wait_and_tunnel():
    if check_server_ready(PORT):
        print("ComfyUI ready!")
        start_tunnel(PORT)
    else:
        print("Server did not start in time")

threading.Thread(target=wait_and_tunnel, daemon=True).start()

# Start ComfyUI (blocking)
%cd /content/ComfyUI
!python main.py $EXTRA_ARGS
```
